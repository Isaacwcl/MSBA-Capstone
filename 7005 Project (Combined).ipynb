```python
df4
```

```python
all_df = pd.concat([df1, df2, df3, df4], axis=0)
```

```python
all_df = all_df.reset_index()
```

```python
## Define member or non-member with a new column "MEMBER_2"

member = []
for i in range(len(all_df)):
    if pd.isna(all_df["MEMBER"][i]) == True:
        member.append("Non-member")
    else:
        member.append("Member")
```

```python
all_df["MEMBER_2"] = member
```

```python
## Define PB or non-PB sales with a new column "PB_2"

pb = []
for i in range(len(all_df)):
    if all_df['PRODUCT'][i].startswith('TV') == True:
        pb.append("PB")
    else:
        pb.append("Non-PB")
```

```python
all_df["PB_2"] = pb
```

```python
## Calculating the sum of product sales for each transaction

sales = []
for i in range(len(all_df)):
    product_sales = all_df['PRICE'][i]*all_df['QUANTITY'][i]
    sales.append(product_sales)
```

```python
all_df["SALES"] = sales
```

## 2.3 Membership rate

### 2.3.1 Overall memebership rate

```python
# Number of memebers and non-members buying PB goods

num_mem = PB_data.loc[PB_data["MEMBER"].notnull(), "INVOICE"].nunique()
num_nonmem = PB_data.loc[PB_data["MEMBER"].isnull(), "INVOICE"].nunique()
pd.DataFrame({"Category": ["Member", "Non-member"], "Count": [num_mem, num_nonmem]})
```

```python
# Count the number of unique invoices with a non-null "MEMBER" value
num_members = PB_data.loc[PB_data["MEMBER"].notnull(), "INVOICE"].nunique()

# Count the total number of unique invoices
total_invoices = PB_data["INVOICE"].nunique()

# Calculate the membership rate
membership_rate = (num_members / total_invoices) * 100

# Print the membership rate
print(f"Membership rate: {membership_rate:.2f}%")
```

```python
avg_mem = PB_data.loc[PB_data["MEMBER_2"] == "Member"]["SALES"].sum() / num_mem
print(f"The average PB sales of member is {avg_mem}.")
```

```python
avg_nonmem = PB_data.loc[PB_data["MEMBER_2"] == "Non-member"]["SALES"].sum() / num_nonmem
print(f"The average PB sales of non-member is {avg_nonmem}.")
```

### T-test

```python
std1 = np.std(PB_data.loc[PB_data["MEMBER_2"] == "Member", "SALES"])
std2 = np.std(PB_data.loc[PB_data["MEMBER_2"] == "Non-member", "SALES"])
```

```python
# Perform the t-test
t_stat, p_value = stats.ttest_ind_from_stats(avg_mem, std1, num_mem, avg_nonmem, std2, num_nonmem)

# Print the t-statistic and p-value
print("T-statistic:", t_stat)
print("P-value:", p_value)
```

From the t-test, there is a statistical difference in PB sales between members and non-members at 5% significance level. Member will be contributing to higher PB sales.

### T-test

```python
def multi_cm(x):
    num_mem = PB_data.loc[(PB_data["MEMBER_2"] == "Member") & (PB_data['STORE'] == x), "INVOICE"].nunique()
    num_nonmem = PB_data.loc[(PB_data["MEMBER_2"] == "Non-member") & (PB_data['STORE'] == x), "INVOICE"].nunique()
    avg_mem = PB_data.loc[(PB_data["MEMBER_2"] == "Member") & (PB_data['STORE'] == x)]["SALES"].sum() / num_mem
    avg_nonmem = PB_data.loc[(PB_data["MEMBER_2"] == "Non-member") & (PB_data['STORE'] == x)]["SALES"].sum() / num_nonmem
    std1 = np.std(PB_data.loc[(PB_data["MEMBER_2"] == "Member") & (PB_data['STORE'] == x), "SALES"])
    std2 = np.std(PB_data.loc[(PB_data["MEMBER_2"] == "Non-member") & (PB_data['STORE'] == x), "SALES"])
    t_stat, p_value = stats.ttest_ind_from_stats(avg_mem, std1, num_mem, avg_nonmem, std2, num_nonmem)
    cprint(f"STORE {x}",'red', 'on_yellow')
    print("T-statistic:", t_stat)
    print("P-value:", p_value) 
```

```python
for store in range(1,13):
    multi_cm(store)
```

Store 1, 2, 3, 4 and 6 showed statistical difference in PB sales between member and non-member at 5% significance lvel. In these stores, members contributes to higher PB sales.
