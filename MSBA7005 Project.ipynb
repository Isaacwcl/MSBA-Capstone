```python
import pandas as pd
```

```python
df = pd.read_excel('NS_Peak_1.xlsx', sheet_name='商品明細報表', skiprows=1)
```

```python
df
```

```python
sales = []
for i in range(len(df)):
    product_sales = df['PRICE'][i]*df['QUANTITY'][i]
    sales.append(product_sales)
```

```python
df["SALES"] = sales
```

```python
PB_data = df[df['PRODUCT'].str.startswith('TV')]
```

```python
PB_data
```

```python
# No. of PB sales in different stores
PB_data["STORE"].value_counts().sort_index()
```

```python
# No. of PB sales in different categories
PB_data["DEPARTMENT"].value_counts()
```

```python
# PB sales in different categories
PB_data.groupby("DEPARTMENT")["SALES"].sum().sort_values(ascending=False)
```

```python
PB_data.loc[PB_data["DEPARTMENT"] == "HOUSEHOLD & EA"]["SECTION NAME"].value_counts()
```
