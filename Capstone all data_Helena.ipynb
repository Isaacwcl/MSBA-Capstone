## Overall analysis

```python
data_frames = [Non_sale_Peak, Non_sale_Non_peak, Sale_Peak, Sale_Non_peak]

# ratio of member and nonmember
for df, name in zip(data_frames, ['Non_sale_Peak', 'Non_sale_Non_peak', 'Sale_Peak', 'Sale_Non_peak']):
    member_transactions = df[df['member_id'] != 'nonmember']
    non_member_transactions = df[df['member_id'] == 'nonmember']
    print(f"{name}:")
    print(f"Ratio of member transaction: {len(member_transactions) / len(df) * 100:.2f}%")
    print(f"Ratio of nonmember transaction: {len(non_member_transactions) / len(df) * 100:.2f}%")
    print()
```

```python
# PB and NonPB
for df, name in zip(data_frames, ['Non_sale_Peak', 'Non_sale_Non_peak', 'Sale_Peak', 'Sale_Non_peak']):
    # PB items
    pb_items = df[df['product_name'].str.contains('TV|TVHC', case=False, na=False)]
    non_pb_items = df[~df['product_name'].str.contains('TV|TVHC', case=False, na=False)]
    
    total_items = len(df)
    total_sales_amount = df['sales_price'].sum()

    print(f"{name}:")
    # PB items
    pb_item_count = len(pb_items)
    pb_sales_amount = pb_items['sales_price'].sum()
    pb_item_percentage = (pb_item_count / total_items) * 100 if total_items != 0 else 0
    pb_sales_percentage = (pb_sales_amount / total_sales_amount) * 100 if total_sales_amount != 0 else 0
    print(f"Number of PB items: {pb_item_count}")
    print(f"Total sales amount of PB items: {pb_sales_amount}")
    print(f"Percentage of PB items: {pb_item_percentage:.2f}%")
    print(f"Percentage of sales amount of PB items: {pb_sales_percentage:.2f}%")
    
    # Non-PB items
    non_pb_item_count = len(non_pb_items)
    non_pb_sales_amount = non_pb_items['sales_price'].sum()
    non_pb_item_percentage = (non_pb_item_count / total_items) * 100 if total_items != 0 else 0
    non_pb_sales_percentage = (non_pb_sales_amount / total_sales_amount) * 100 if total_sales_amount != 0 else 0
    print(f"Number of Non-PB items: {non_pb_item_count}")
    print(f"Total sales amount of Non-PB items: {non_pb_sales_amount}")
    print(f"Percentage of Non-PB items: {non_pb_item_percentage:.2f}%")
    print(f"Percentage of sales amount of Non-PB items: {non_pb_sales_percentage:.2f}%")
    print()
```

```python
# PB and NonPB for members
for df, name in zip(data_frames, ['Non_sale_Peak', 'Non_sale_Non_peak', 'Sale_Peak', 'Sale_Non_peak']):
    
    member_transactions = df[df['member_id'] != 'nonmember']
    
    # PB items for members
    pb_member_items = member_transactions[member_transactions['product_name'].str.contains('TV|TVHC', case=False, na=False)]
    non_pb_member_items = member_transactions[~member_transactions['product_name'].str.contains('TV|TVHC', case=False, na=False)]
    
    total_member_items = len(member_transactions)
    total_member_sales_amount = member_transactions['sales_price'].sum()

    print(f"{name} (Members):")
    
    # PB items for members
    pb_member_item_count = len(pb_member_items)
    pb_member_sales_amount = pb_member_items['sales_price'].sum()
    pb_member_item_percentage = (pb_member_item_count / total_member_items) * 100 if total_member_items != 0 else 0
    pb_member_sales_percentage = (pb_member_sales_amount / total_member_sales_amount) * 100 if total_member_sales_amount != 0 else 0
    print(f"Number of PB items (Members): {pb_member_item_count}")
    print(f"Total sales amount of PB items (Members): {pb_member_sales_amount}")
    print(f"Percentage of PB items (Members): {pb_member_item_percentage:.2f}%")
    print(f"Percentage of sales amount of PB items (Members): {pb_member_sales_percentage:.2f}%")
    
    # Non-PB items for members
    non_pb_member_item_count = len(non_pb_member_items)
    non_pb_member_sales_amount = non_pb_member_items['sales_price'].sum()
    non_pb_member_item_percentage = (non_pb_member_item_count / total_member_items) * 100 if total_member_items != 0 else 0
    non_pb_member_sales_percentage = (non_pb_member_sales_amount / total_member_sales_amount) * 100 if total_member_sales_amount != 0 else 0
    print(f"Number of Non-PB items (Members): {non_pb_member_item_count}")
    print(f"Total sales amount of Non-PB items (Members): {non_pb_member_sales_amount}")
    print(f"Percentage of Non-PB items (Members): {non_pb_member_item_percentage:.2f}%")
    print(f"Percentage of sales amount of Non-PB items (Members): {non_pb_member_sales_percentage:.2f}%")
    print()
```

```python
# PB and NonPB for nonmembers
for df, name in zip(data_frames, ['Non_sale_Peak', 'Non_sale_Non_peak', 'Sale_Peak', 'Sale_Non_peak']):
    
    nonmember_transactions = df[df['member_id'] == 'nonmember']
    
    # PB items for members
    pb_nonmember_items = nonmember_transactions[nonmember_transactions['product_name'].str.contains('TV|TVHC', case=False, na=False)]
    non_pb_nonmember_items = nonmember_transactions[~nonmember_transactions['product_name'].str.contains('TV|TVHC', case=False, na=False)]
    
    total_nonmember_items = len(nonmember_transactions)
    total_nonmember_sales_amount = nonmember_transactions['sales_price'].sum()

    print(f"{name} (Nonembers):")
    
    # PB items for nonmembers
    pb_nonmember_item_count = len(pb_nonmember_items)
    pb_nonmember_sales_amount = pb_nonmember_items['sales_price'].sum()
    pb_nonmember_item_percentage = (pb_nonmember_item_count / total_nonmember_items) * 100 if total_nonmember_items != 0 else 0
    pb_nonmember_sales_percentage = (pb_nonmember_sales_amount / total_nonmember_sales_amount) * 100 if total_nonmember_sales_amount != 0 else 0
    print(f"Number of PB items (Nonmembers): {pb_nonmember_item_count}")
    print(f"Total sales amount of PB items (Nonmembers): {pb_nonmember_sales_amount}")
    print(f"Percentage of PB items (Nonmembers): {pb_nonmember_item_percentage:.2f}%")
    print(f"Percentage of sales amount of PB items (Nonmembers): {pb_nonmember_sales_percentage:.2f}%")
    
    # Non-PB items for nonmembers
    non_pb_nonmember_item_count = len(non_pb_nonmember_items)
    non_pb_nonmember_sales_amount = non_pb_nonmember_items['sales_price'].sum()
    non_pb_nonmember_item_percentage = (non_pb_nonmember_item_count / total_nonmember_items) * 100 if total_nonmember_items != 0 else 0
    non_pb_nonmember_sales_percentage = (non_pb_nonmember_sales_amount / total_nonmember_sales_amount) * 100 if total_nonmember_sales_amount != 0 else 0
    print(f"Number of Non-PB items (Nonmembers): {non_pb_nonmember_item_count}")
    print(f"Total sales amount of Non-PB items (Nonmembers): {non_pb_nonmember_sales_amount}")
    print(f"Percentage of Non-PB items (Nonmembers): {non_pb_nonmember_item_percentage:.2f}%")
    print(f"Percentage of sales amount of Non-PB items (Nonmembers): {non_pb_nonmember_sales_percentage:.2f}%")
    print()
```

第四个数据集1911_1011（SaleDay & NonPeakHour）存在两点异常：第一，会员交易量占比极低，其余三个数据集会员交易数量所占比为30%~50%，这个数据集会员交易量只有2.6%；第二，该数据集记录的会员交易中，所购买的产品全部为PB产品，不符实际。

**总体销售情况：**

1.促销活动可以显著提高所有产品的销售量，尤其是在高峰时段（注：但利润未知）；

2.不管是高峰时段还是非高峰时段，促销日中PB产品的销售量占比有显著增加，说明促销日是促进PB产品销售的有效方式。


**会员购买行为：**

1.非会员对于PB产品的购买占比更大，表明PB产品可能不是吸引会员的主要因素。


**策略建议：**

1.加强促销活动

*1.1 高峰时段和促销日安排更多促销活动，尤其是针对PB产品的促销活动（有可能现在的促销已经加强了对PB产品的促销，因此导致了“总体销售情况”中第二点的发生，在这种情况下此建议可能不再适用）；*

*1.2 细化对会员和非会员的促销活动，提供专属优惠积分奖励（一会搜索一下现在的优惠政策），进一步提升PB产品的销售。*

2.会员营销

*2.1 针对会员推出定期的促销活动和特别优惠，以增强会员的忠诚度和购买频率；*

*2.2 利用会员数据进行精准营销，推送个性化的商品推荐和促销信息，提高会员的参与度和购买意愿。*
